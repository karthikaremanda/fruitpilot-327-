```python
from ultralytics import YOLO
import cv2

# Load a pretrained YOLO model
model = YOLO("yolov8n.pt")  # Using YOLOv8 nano for faster inference

# List of images to process
image_paths = ["image1.jpg", "image2.jpg", "image3.jpg"]  # Change this to your actual images

for img_path in image_paths:
    results = model(img_path)  # Run object detection
    detections = results[0].boxes.xywhn  # Get bounding boxes (normalized format)
    classes = results[0].boxes.cls  # Get class IDs

    # Save detections in YOLO format
    label_file = img_path.replace(".jpg", ".txt")
    with open(label_file, "w") as f:
        for box, cls in zip(detections, classes):
            x_center, y_center, width, height = box.tolist()
            f.write(f"{int(cls)} {x_center} {y_center} {width} {height}\n")

print("Auto-labeling complete! Labels saved as .txt files.")
```

```python
import torch
import os
import cv2

# Path to your dataset
image_dir = '/content/dataset/images/train'
label_dir = '/content/dataset/labels/train'

# Load the pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # You can use a different model like 'yolov5m'

# Create label directory if it doesn't exist
os.makedirs(label_dir, exist_ok=True)

# Process each image in the dataset
for image_name in os.listdir(image_dir):
    # Skip non-image files
    if not image_name.endswith(('.jpg', '.png', '.jpeg')):
        continue

    # Load the image
    image_path = os.path.join(image_dir, image_name)
    image = cv2.imread(image_path)

    # Get model predictions
    results = model(image)

    # Extract bounding boxes and class predictions
    boxes = results.xywh[0].numpy()  # [x_center, y_center, width, height]
    class_ids = results.names  # class names list

    # Write the label file
    label_file_path = os.path.join(label_dir, image_name.replace('.jpg', '.txt').replace('.png', '.txt'))

    with open(label_file_path, 'w') as label_file:
        for box in boxes:
            # Get class ID and coordinates
            class_id = int(box[5])  # the last element is the class id
            center_x, center_y, width, height = box[:4]

            # Write in YOLO format (normalized)
            label_file.write(f"{class_id} {center_x} {center_y} {width} {height}\n")

    print(f"Processed {image_name}")
```
